### Onehot encoding to catdata

```python
catdata = data[['Material','Weather_Conditions','Material_Composition','Bridge_Design']]
```

```python
catdata.head(3)
```

```python
# catdata_final.head(3)
```

```python
# catdata
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Define categorical columns
cat_cols = ['Material', 'Weather_Conditions','Material_Composition','Bridge_Design']

# Apply One-Hot Encoding
encoder = OneHotEncoder(sparse_output=False)
encoded_array = encoder.fit_transform(catdata[cat_cols])

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(cat_cols))

# Merge with original dataset (drop original categorical columns)
cat_final = catdata.drop(columns=cat_cols).reset_index(drop=True)
cat_final = pd.concat([cat_final, encoded_df], axis=1)

print(cat_final.head())
```

```python
cat_final.head(3)
```

### PAC to numdata

```python
numdata = data[['Age (years)','Length (m)','Width (m)','Height (m)','Traffic_Volume (vehicles/day)','Water_Flow_Rate (m³/s)','Stress (MPa)','Strain (%)','Tensile_Strength (MPa)','Rainfall (mm)','Temperature (°C)','Humidity (%)','Collapse_Status']]
```

```python
numdata.head(3)
```

```python
x = numdata.drop("Collapse_Status", axis=1) ##Here we drop the output features and only keep the input features
y = numdata["Collapse_Status"]  ##Here we shows only the output features
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 🔹 Load dataset (Replace with your actual dataset)
# df = pd.read_csv("your_dataset.csv")

# 🔹 Drop non-numeric columns (like categorical) if not encoded yet
x_numeric = x.select_dtypes(include=[np.number])

# 🔹 Standardize the data (PCA works best with scaled data)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_numeric)

# 🔹 Apply PCA
pca = PCA(n_components=0.95)  # Keep 95% variance
pca.fit(x_scaled)

# 🔹 Get the importance of each original feature in PCA components
importance = np.abs(pca.components_).sum(axis=0)

# 🔹 Create a DataFrame to display feature importance
feature_importance = pd.DataFrame({'Feature': x_numeric.columns, 'Importance': importance})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

# 🔹 Print the most important features
print("Top Important Features:")
print(feature_importance.head(10))  # Show top 10 features
```

```python
num_final= numdata[['Age (years)','Length (m)','Humidity (%)','Traffic_Volume (vehicles/day)','Temperature (°C)','Width (m)','Height (m)','Water_Flow_Rate (m³/s)','Stress (MPa)','Strain (%)']]
```

```python
# Here we import split function from sklearn
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
for col in X_train.columns:
    plt.figure(figsize=(10,10))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)

    
    plt.figure(figsize=(10,10))
    plt.subplot(122)
    stats.probplot(X_train[col], dist='norm',plot=plt)
    plt.title(col)
    plt.show()
```

```python
X_train_transformed = scaler.fit_transform(X_train)
X_test_transoformed = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score
from math import sqrt
```

```python
# Define the regression models
models = {
    'lin_reg': LinearRegression(),
    'rf_reg': RandomForestRegressor(n_estimators=100, random_state=42),
    'dt_reg' : tree.DecisionTreeClassifier()
    
}
```

```python
for name ,model in models.items():
    model.fit(X_train_transformed,y_train)
    y_pred = model.predict(X_test_transoformed)
    
    print(f"{name} : mae:{mean_absolute_error(y_test,y_pred)}, mse :{mean_squared_error(y_test,y_pred)}, rmse :{root_mean_squared_error(y_test,y_pred)}, r2_score : {r2_score(y_test,y_pred)}")
```
